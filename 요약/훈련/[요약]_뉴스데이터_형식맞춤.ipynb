```python
from typing import List
from konlpy.tag import Okt

class OktTokenizer:
    okt: Okt = Okt()

    def __call__(self, text: str) -> List[str]:
        tokens: List[str] = self.okt.phrases(text)
        return tokens
```

# 1. 구글

```python
google = pd.read_csv('/content/drive/MyDrive/miraeasset/data/[구글]1-15페이지_완성(찐).csv')
google
```

## 중복 확인 및 제거

```python
# 한 공모주 내 중복 있는지 살피기
google.drop_duplicates(['title', 'name'], keep='first', inplace=True)
google.reset_index(drop=True, inplace=True)
google
```

```python
# 기사 내 해당 공모주명 없는 데이터 삭제
google = google[google.apply(lambda row: row['name'] in row['content'], axis=1)]
google
```

## content 길이 다듬기 -> textrank

### 텍스트 개수와 문장 개수 세기

```python
import re
google.content = google.content.apply(lambda x: re.sub(' +',' ',x))
google.content = google.content.apply(lambda x: re.sub('\n','',x))
google.content = google.content.apply(lambda x:re.sub('\u200b','',x))
google.content = google.content.apply(lambda x:re.sub('\xa0','',x))
google.content = google.content.apply(lambda x:re.sub('[-=+,#/\?:^$@*\"※~&%ㆍ!』\\‘|\(\)\[\]\<\>`\'…》]','',x))
```

```python
text_len=[]
for text in google.content:
  length=len(text.split())
  text_len.append(length)
```

```python
google['text_len']=text_len
```

```python
google.sort_values(by=['text_len'],ascending=False)
```

```python
# 문장 갯수
sent_count=[]
for sent in google['content']:
  sentences= re.split(r'[.!?]', sent)
  non_empty_sentences = [sentence.strip() for sentence in sentences if sentence.strip()]
  sent_count.append(len(non_empty_sentences))
google['sent_count']=sent_count
google.sort_values(by=['sent_count'],ascending=False)
```

```python
from typing import List
from textrankr import TextRank

mytokenizer: OktTokenizer = OktTokenizer()
textrank: TextRank = TextRank(mytokenizer)

k: int =20 #20문장으로 요약
summaries: List[str] = []

for index, row in google.iterrows():
    text = row['content']
    if row['sent_count']>20: #문장개수 40개 이상만 요약하기
      summary = textrank.summarize(text, k)
      summaries.append(summary)
    else:
      summaries.append(text)

# 요약 결과를 'short' 칼럼에 담기
google['short'] = summaries

google.reset_index(drop=True, inplace=True)
google
```

## column 정리

```python
# 중간저장
google.to_csv('/content/drive/MyDrive/miraeasset/data/[구글]IPO_news0.csv', index=False)
```

```python
# google link 칼럼 제거
google = google.drop(columns=['link', 'text_len','sent_count','content'])
google
```

```python
google = google.rename(columns={'short':'content'})
google
```

## 특수기호 제거

```python
# 특수문자를 제외한 알파벳, 숫자, 한글만 남기는 함수
def clean_text(text):
    cleaned = re.sub(r'[^a-zA-Z0-9가-힣ㄱ-ㅎㅏ-ㅣ ,.!?]', '', text)
    return cleaned
```

```python
# 데이터프레임 복사
filtered_g = google.copy()
filtered_g
```

```python
#결측값 처리
filtered_g = filtered_g[~filtered_g['title'].isna()]
filtered_g.reset_index(drop=True, inplace=True)
filtered_g
```

```python
filtered_g['title'] = filtered_g['title'].apply(clean_text)
filtered_g['content'] = filtered_g['content'].apply(clean_text)
filtered_g
```

```python
filtered_g.to_csv('/content/drive/MyDrive/miraeasset/data/[구글]IPO_news.csv', index=False)
```
