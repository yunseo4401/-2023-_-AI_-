```python
def get_completion(name, prompt, model="gpt-3.5-turbo", temperature=0.2):
    '''
    name : 해당 공모주명
    prompt: 해당 프롬프트 -> 기사
    model: 사용할 모델, 기본값은 gpt-3.5-turbo(ChatGPT)
    temperature: 온도 계수
    '''
    messages = [
        {'role': 'system', 'content': '"'+name+'"에 대한 뉴스 기사를 최대 200자 이내로 요약해줘.'},
        {"role": "user", "content": prompt}
        ]
    response = openai.ChatCompletion.create(
        model=model,
        messages=messages,
        temperature=temperature, # 모델 출력의 온도 계수, 출력의 무작위성을 제어합니다.
        max_tokens=1000
    )
    # OpenAI의 ChatCompletion 인터페이스를 호출.
    return response.choices[0].message["content"]
```

```python
# 한글 포함 여부를 체크하는 함수 정의
def contains_hangul(s):
    return any(c for c in s if '가' <= c <= '힣')

# 한글이 하나도 포함되어 있지 않은 row 필터링
filtered_df = concat[~concat['result'].apply(contains_hangul)]
filtered_df
```
