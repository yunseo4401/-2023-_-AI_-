```python
from tqdm import tqdm
import numpy as np
import time
def trans_to_kor(trans_list, text_data,transed_lang,start_index,final_index): #영어를 한글로 번역하는 함수
  driver = webdriver.Chrome(options=options)
  driver.get(base_url)
  trans_list=trans_list
  target_present = EC.presence_of_element_located((By.XPATH, '//*[@id="txtTarget"]'))

  for i in tqdm(range(start_index,final_index)):

    if (i!=0)&(i%99==0):
      time.sleep(1.10)

    if text_data[i]=='':
      trans_list.append('')
      continue

    try:
      driver.get('https://papago.naver.com/?sk=en&tk='+transed_lang+'&st='+text_data[i])
      time.sleep(10)
      element=WebDriverWait(driver, 10).until(target_present)
      time.sleep(0.10)
      backtrans = element.text

      if (backtrans=='')|(backtrans==' '):
        element=WebDriverWait(driver, 10).until(target_present)
        backtrans = element.text
        trans_list.append(backtrans)
      else:
        trans_list.append(backtrans)

    except:
      trans_list.append('')
```
