```python
# 모델 및 토크나이저 불러오기
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

tokenizer = AutoTokenizer.from_pretrained("Sehong/kobart-QuestionGeneration")
model = AutoModelForSeq2SeqLM.from_pretrained("Sehong/kobart-QuestionGeneration")
```

```python
q_list=[] # 생성된 질문 저장할 리스트
```

```python
import torch

# 기사 내용 하나씩 가져와서 모델에 입력하기. -> 질문 출력 -> q_list에 append
for i in range(len(data)):
  text=data['content'][i]
  raw_input_ids = tokenizer.encode(text)
  input_ids = [tokenizer.bos_token_id] + raw_input_ids + [tokenizer.eos_token_id]
  try:
    summary_ids = model.generate(torch.tensor([input_ids]),max_length= 60)
    print(tokenizer.decode(summary_ids.squeeze().tolist(), skip_special_tokens=True))
    q_list.append(tokenizer.decode(summary_ids.squeeze().tolist(), skip_special_tokens=True))
  except:
    q_list.append('')
```

```python
# 중복 제거, 이상한 문장 제거, 물음표 없는 것 제거, 인덱스는 없애지 말기.
```

```python
# 길이 파악
print(len(q_list), len(set(q_list)))
```

```python
df=pd.DataFrame(columns=['context','question'])
```

```python
df['context'] = data['content'].copy()
df['question']=q_list

df
```

```python
# 생성된 질문이 중복된 행 삭제
df.drop_duplicates(['question'], inplace=True)
```

```python
df
```

```python
# 물음표가 하나도 포함되지 않은 row 필터링하는 함수 정의
def does_not_contain_question_mark(s):
    return '?' not in s


# 물음표가 하나도 포함되지 않은 row 필터링
filtered_df = df[~df['question'].apply(lambda row: all(map(does_not_contain_question_mark, row)))]
filtered_df
```

```python
filtered_df.to_csv('kobart_result.csv')
```

```python
def get_completion(context, question, model="gpt-3.5-turbo", temperature=0.2):
    '''
    context : 해당 기사 원문
    question: 해당 프롬프트 -> 기사
    model: 사용할 모델, 기본값은 gpt-3.5-turbo(ChatGPT)
    temperature: 온도 계수
    '''
    messages = [
        {'role': 'system', 'content': context},
        {"role": "user", "content": question}
        ]
    response = openai.ChatCompletion.create(
        model=model,
        messages=messages,
        temperature=temperature, # 모델 출력의 온도 계수, 출력의 무작위성을 제어
        max_tokens=1000
    )
    # OpenAI의 ChatCompletion 인터페이스를 호출한다.
    return response.choices[0].message["content"]
```
