## 지식인 정제 데이터

```python
# 지식인 정제 파일 모으기
import pandas as pd

data1=pd.read_csv('/content/drive/MyDrive/미래에셋/mypart (1) (3).csv',encoding='CP949') #1
data2=pd.read_csv('/content/drive/MyDrive/미래에셋/chanwoo_only_qna.csv',encoding='CP949') #2
```

```python
# 각 데이터에서 정제된 답변 칼럼 가져오기
data1=data1[['revise']]
data2=data2[['answer']]
data2=data2.dropna() # 결측치 제거
```

```python
data1=data1.rename(columns={'revise':'answer'}) # 칼럼명 data2와 통일하기
```

```python
nogada=pd.concat([data1,data2],axis=0) # 두 데이터 병합
nogada
```

## 아하 정제 데이터

```python
import pandas as pd

data3=pd.read_csv('/content/drive/MyDrive/미래에셋/아하_zzin_ cw.csv',encoding='CP949') #아하1
data4=pd.read_csv('/content/drive/MyDrive/미래에셋/아하_zzin_ys.csv',encoding='CP949') #아하2
data5=pd.read_csv('/content/drive/MyDrive/미래에셋/아하_0to450.csv',encoding='CP949')# 아하3
```

```python
# 각 데이터의 결측치 및 인덱스 제거

data3=data3.dropna()
data3.reset_index(inplace=True)
data4=data4.dropna()
data4.reset_index(inplace=True)
data5=data5.dropna()
data5.reset_index(inplace=True)
```

```python
aha=pd.concat([data3,data4,data5],axis=0) #아하 합치기
aha
```

```python
aha=aha[['answer']]
aha=aha.dropna()
```

```python
nogada=pd.concat([nogada,aha],axis=0) #아하랑 지식인 정제 데이터 합치기
```
