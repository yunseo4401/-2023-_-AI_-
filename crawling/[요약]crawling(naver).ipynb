```python
def convert_relative_date(relative_date):
  """들어온 날짜 데이터에 문자열이 포함되어 있다면 날짜 형식으로 변환해준다.

    Args:
        relative_date: 기사 작성일.

    Returns:
        통일된 날짜 형식으로 반환.
    """
  if '어제' in relative_date:
    current_date = datetime.today()
    modified_date = current_date - timedelta(days=1)
    return modified_date.strftime('%Y. %m. %d.')

  elif '일 전' in relative_date:
    days_ago = int(relative_date.split('일 전')[0])
    current_date = datetime.today()
    modified_date = current_date - timedelta(days=days_ago)
    return modified_date.strftime('%Y. %m. %d.')

  elif '시간 전' in relative_date:
    hours_ago = int(relative_date.split('시간 전')[0])
    current_date = datetime.today()
    modified_d
    ate = current_date - timedelta(hours=hours_ago)
    return modified_date.strftime('%Y. %m. %d.')

  else:
    return relative_date
```

```python
# real_result : df에서

real_result = pd.DataFrame(columns=['link', 'article', 'name', 'title', 'date'])

for i in range(15):
    page = globals()[f"page_{i+1}"] # 한 페이지 당 데이터 프레임
    ipo_list = globals()[f"ipo_list_{i+1}"] # 공모주 리스트
    result_list = globals()[f"result_list_{i+1}"] # 청약일 (D-1, D-30) 리스트

    # 공모주별 날짜 데이터 변환 후 날짜 필터링 & 데이터프레임 누적
    for k in range(len(result_list)):
        df = page[page['name'] == ipo_list[k]].copy()
        start_date = result_list[k][0]  # 청약일 D-1
        end_date = result_list[k][1]  # 청약일 D-30

        # 날짜 데이터 변환
        df['date'] = df['date'].apply(convert_relative_date)
        df['date'] = pd.to_datetime(df['date'], format='%Y.%m.%d.')
        # 기사 작성일 D-1 ~ D-30 안에 들어가는 것만 필터링
        df = df[(df['date'] >= end_date) & (df['date'] <= start_date)]

        real_result = pd.concat([real_result, df], ignore_index=True) # 전체데이터에 현재 페이지 데이터 병합

real_result
```

## 1. 기사 링크 리스트 크롤링

- 구글 뉴스에서 `'"각 공모주명"+"공모주"'`를 키워드로 설정하여 검색한 결과 속 각 신문사로 이어지는 뉴스 링크 리스트를 크롤링 하였습니다.
- `real_result` : 기사별 링크(link), 신문사(article), 관련 공모주(name), 날짜(date) 데이터 **프레임**

```python
def naver(link):
    """
    네이버 뉴스 페이지를 크롤링하여 기사 링크, 제목, 기사 내용, 날짜 정보를 추출하는 함수.

    Args:
        link (str): 크롤링할 웹 페이지의 링크 URL.

    Returns:
        tuple: 기사 링크, 기사 내용, 제목, 날짜 정보가 담긴 튜플.
    """

    # 주어진 링크로 HTTP GET 요청을 보냄
    response = requests.get(link)

    # 인코딩을 UTF-8로 설정하여 한글과 같은 문자를 올바르게 처리
    response.encoding = 'utf-8'

    # BeautifulSoup를 사용하여 HTML 문서 파싱
    soup = BeautifulSoup(response.text, 'html.parser')

    # 기사 링크, 제목, 기사 내용, 날짜를 저장할 리스트 초기화
    article_link = []
    title_list = []
    article_list = []
    date_list = []

    # HTML 문서에서 클래스가 'news_tit'인 링크 요소를 찾아서 최대 10개까지 가져옴
    news_links = soup.find_all('a', attrs={'class': 'news_tit'})
    for i in range(min(10, len(news_links))):
        article_link.append(news_links[i].get('href'))
        title_list.append(news_links[i].get('title'))

    # 기사 내용을 가져옴
    for i in range(min(10, len(news_links))):
        art_temp = soup.find_all('a', attrs={'class': 'info press'})[i]
        # 기사 내용에서 'i' 태그를 제거
        if art_temp.find('i') is not None:
            art_temp.find('i').decompose()
        article_list.append(art_temp.text)

    # 기사의 날짜 정보를 가져옴
    date = soup.find_all('span', attrs={'class': 'info'})
    for i in range(min(20, len(date))):
        date_temp = soup.find_all('span', attrs={'class': 'info'})[i]

        # '면'이 포함된 텍스트는 무시하고 날짜 리스트에 추가
        word_to_remove = '면'
        if word_to_remove in date_temp.text:
            continue
        date_list.append(date_temp.text)

    # 기사 링크, 기사 내용, 제목, 날짜 리스트를 반환
    return article_link, article_list, title_list, date_list
```

```python
retry_list = []
df = pd.DataFrame(columns=['link', 'article', 'name', 'title', 'date'])

crawling = True
for h in range(15):
  cnt = 0 # 한 페이지마다 cnt = 0으로 초기화
  globals()[f"page_df_{h+1}"] = pd.DataFrame(columns=['link', 'article', 'name', 'title', 'date'])

  # 현재 페이지의 IPO 목록을 순회
  for k in range(len(globals()[f"ipo_list_{h+1}"])):
    stock = globals()[f"ipo_list_{h+1}"][k] # 공모주 이름 가져오기
    print(f"{stock} 크롤링 시작 !@!")
    keyword =f'"{stock}"%2B"공모주"' # 검색을 위해 주식 이름을 따옴표로 감싸기
    max_page = 100 # 크롤링할 최대 페이지 수

    # 페이지 순회
    for j in range(max_page):
        try:
            if j == 0:
                url = "https://search.naver.com/search.naver?where=news&sm=tab_pge&query="+keyword+"&sort=0&photo=0&field=0&pd=0&ds=&de=&cluster_rank=19&mynews=0&office_type=0&office_section_code=0&news_office_checked=&nso=so:r,p:all,a:all&start=1"
            else:
                url = "https://search.naver.com/search.naver?where=news&sm=tab_pge&query="+keyword+f"&sort=0&photo=0&field=0&pd=0&ds=&de=&cluster_rank=19&mynews=0&office_type=0&office_section_code=0&news_office_checked=&nso=so:r,p:all,a:all&start={j}1"

            naver_data = naver(url)  # 기사 내용 크롤링
            if len(naver_data[0]) == 0 or len(naver_data[1]) == 0 or len(naver_data[2]) == 0 or len(naver_data[3]) == 0:
              print(f"{stock} 크롤링 다시 해봐. 기사 0개야.")
              retry_list.append(stock) # 크롤링 안 된 주식 이름 저장
              print(f"{stock}(은)는 {h+1}페이지의 {k+1}번째 공모주야.")
              crawling = False
              if cnt < 30: # 한 페이지에 크롤링 안 된게 30개 이상이라면, 멈춘다.
                cnt += 1 # 여기 들어오면 cnt 증가
                print(f"*** {h+1}페이지에서 {cnt}번 크롤링 안됨 ***")
                crawling = True # 넘기기
                break
              elif crawling == False:
                break

            if len(naver_data[0]) < 10 or len(naver_data[1]) < 10 or len(naver_data[2]) < 10 or len(naver_data[3]) < 10:
                # 해당 페이지에 기사가 10개 미만이라면, 마지막 페이지라는 뜻.
                print(f"Reached end of data on page {j + 1}")
                # 남아있는 만큼 크롤링.
                for i in range(len(naver_data[0])):
                  df = df.append({
                      'link': naver_data[0][i],
                      'article': naver_data[1][i],
                      'name': f'{stock}',
                      'title': naver_data[2][i],
                      'date': naver_data[3][i]
                  }, ignore_index=True)
                break
            # 얘네들은 모두 한 페이지에 기사가 10개 있는 페이지들.
            for i in range(10):
                df = df.append({
                    'link': naver_data[0][i],
                    'article': naver_data[1][i],
                    'name': f'{stock}',
                    'title': naver_data[2][i],
                    'date': naver_data[3][i]
                }, ignore_index=True)

            time.sleep(3)  # Adding a delay to avoid overloading the server

        except Exception as e:
            # 에러 발생하면 그 페이지 건너뜀.
            print(f"An error occurred on page {j + 1}: {e}")
            print("Retrying after 5 seconds...")
            time.sleep(5)
    if crawling == False:
      break
    print(f"{stock} 크롤링 완료 !@! 기사 개수:", len(df))

  if crawling == False:
      break

  print(f"page_{h+1} 크롤링 완료, 기사 개수:",len(df))

  globals()[f"page_df_{h+1}"] = pd.concat([globals()[f"page_df_{h+1}"], df],ignore_index=True)
  print(f"{h+1} page 데이터프레임 개수:", len(globals()[f"page_df_{h+1}"]))
  df = pd.DataFrame(columns=['link', 'article', 'name', 'title', 'date']) # df 초기화

#  이제 데이터 수집이 완료되었거나 사용 가능한 데이터의 끝에 도달했습니다.
if crawling == False:
  print("Crawling failed.")
else:
  print("Crawling finished.")
```
