## raw data

### load

```python
import pandas as pd
data = pd.read_csv('/content/drive/MyDrive/질문답변쌍.csv')
data2 = pd.read_csv('/content/drive/MyDrive/poly-encoder답변모음.csv')
```

### poly-encoder 쌍

```python
data
```

```python
# 'index' 컬럼을 제거
data = data.drop(columns=['Unnamed: 0'])
data
```

```python
# 열 이름 변경
data = data.rename(columns={'best_answer': 'answer'})
data
```

```python
# 문자열 정리 및 불필요한 문자 제거
data['question'] = data['question'].str.replace(r'[\n\t\s,]+', ' ').str.strip(',')
data
```

### 답변 리스트

```python
data2
```

```python
# 'index' 컬럼을 제거
data2 = data2.drop(columns=['Unnamed: 0'])
data2
```

### 합치기

```python
# 'answer' 열의 각 문자열 값을 리스트로 변환
data['answer'] = data['answer'].apply(lambda x: [x])
```

```python
data
```

```python
data['answer'][0]
```

```python
import random
# 조건에 따라 answer 리스트 확장
for i, row in data.iterrows():
    if len(row['answer']) < len(data2['answer']):
        # 필요한 개수만큼 랜덤한 값 추가
        additional_answers = random.sample(data2['answer'].tolist(), 10 - len(row['answer']))
        data.at[i, 'answer'].extend(additional_answers)
data
```

```python
len(data['answer'][0])
```

```python
import numpy as np
# 리스트를 생성합니다.
label_list = [np.concatenate(([1], np.zeros(len(data2) - 1)))] * len(data)

# label_list 출력
print(label_list)
```

```python
# 소수점을 제거합니다.
label_list = [arr.astype(int) for arr in label_list]
```

```python
label_list[0]
```

```python
len(label_list)
```

```python
len(label_list[0])
```

```python
data['label'] = label_list
```

```python
len(data['answer'][0])
```

## train

```python
train = data[:int(len(data)*0.8)]
```

```python
data
```

```python
# 데이터프레임을 딕셔너리 리스트로 변환
train_data_list = []
for index, row in train.iterrows():
    item = {
        'context': row['question'],
        'labels': row['label'],
        'responses': row['answer'],
    }
    train_data_list.append(item)
```

```python
train_data_list[89]
```

```python
import pickle

# data_list를 pickle 파일로 저장
with open('/content/drive/MyDrive/train.pickle', 'wb') as file:
    pickle.dump(train_data_list, file)
```
